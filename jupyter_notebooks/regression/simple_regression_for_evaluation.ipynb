```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

area = np.array([190, 200, 100, 50, 120]).reshape(-1, 1)
actual = np.array([2100, 1980, 1420, 1100, 1300]).reshape(-1, 1)

model = LinearRegression()
_ = model.fit(area, actual)

fig, ax = plt.subplots(figsize=(4, 4))

y = area * model.coef_ + model.intercept_

_ = ax.plot(area, y, color='tab:blue', linewidth=2, label='hypothesis')
_ = ax.scatter(area, y, color='tab:blue', linewidth=2, label='predicts')

_ = ax.scatter(area, actual, color='tab:red', alpha=0.7, s=50, label='points')

_ = ax.set_ylim([900, 2500])
_ = ax.set_xlim([10, 250])

_ = ax.set_xlabel('area')
_ = ax.set_ylabel('price')

_ = ax.legend(loc='upper left')

# fig.savefig("./figure_01.jpg", dpi=1200, bbox_inches='tight')
```
